```python

from datetime import date


STAGE_COLOR_PALETTE = {
    2 : "firebrick",
    3: "darkgray",
    5: "red",
    6: "yellow",
    7: "darkgreen",
    8 : "mediumslateblue",
    9 : "darkcyan",
    10 : "goldenrod",
    11 : "lime",
    13 : "deepskyblue",
    14 : "cornflowerblue",
    19 : "red",
    21 : "black",
}


def plot_beam_modes(fig,beam_modes,fill_start,fill_stop,set_labels = True):
    """
    Arguments: 
    fig: figure object where to plot
    beam_modes: pandas DataFrame with columns [timestamp, time, cypher, mode]
    fill_start: timestamp indicating fill start
    fill_stop: timestamp indicating fill end
    """
    N = len(beam_modes["time"])


    for i in range(0,N-1):
        if fill_start > beam_modes["timestamp"].iloc[i] or beam_modes["mode"].iloc[i] == "BEAMDUMP" or beam_modes["mode"].iloc[i] == "RAMPDOWN":
            continue
        if set_labels:
            lab = beam_modes["mode"].iloc[i]
        else:
            lab = None
        fig.axvspan(beam_modes["timestamp"].iloc[i],beam_modes["timestamp"].iloc[i+1],
            color = STAGE_COLOR_PALETTE[beam_modes["cypher"].iloc[i]],
            alpha = 0.3, label = lab
            )
    
    if set_labels:
        lab = beam_modes["mode"].iloc[N-1]
    else:
        lab = None
    if beam_modes["mode"].iloc[N-1] == "BEAMDUMP" or beam_modes["mode"].iloc[N-1] == "RAMPDOWN":
        return
    fig.axvspan(beam_modes["timestamp"].iloc[N-1],fill_stop,
        color = STAGE_COLOR_PALETTE[beam_modes["cypher"].iloc[N-1]],
        alpha = 0.3, label = lab
        )




def get_xlabels(timestamps):
    dates = []
    for t in timestamps:
        d = datetime.datetime.fromtimestamp(t)
        dates.append(
            datetime.time(hour = d.hour,
                            minute= d.minute,
                            second= d.second
                                )
        )
    return dates
```

```python
sqrt_s = (bkg.energy_b1["value"].values.max() + bkg.energy_b2["value"].values.max())/1000
sqrt_s
```

```python



frame, fig = plt.subplots(3,1, figsize = (18,12))

fig[0].plot(bkg.times,bkg.lumi,color = "tab:orange",linewidth = 3)
fig0_tw = fig[0].twinx()
fig[0].set_ylabel(r"Instantaneous luminosity Hz/$\mu$b",fontsize = 12)
fig0_tw.plot(bkg.energy_b1["timestamp"],bkg.energy_b1["value"],color = "tab:blue",linewidth = 3)
fig0_tw.set_ylabel("Beam energy [GeV]",fontsize = 12)

plot_beam_modes(fig[0],bkg.beam_mode,bkg.fill_start,bkg.fill_end)
fig[0].legend(bbox_to_anchor = (0.2,0.95))
fig[0].set_title(f"Fill: {bkg.fill_number}, (2022, " + r"$\sqrt{s}$" + " = {:.1f} TeV)".format(sqrt_s) ,
loc = "right", fontsize = 15,)

t0 = np.min(bkg.energy_b1["timestamp"])
t1 = np.max(bkg.energy_b1["timestamp"])
print(bkg.energy_b1["timestamp"].values)
ticks = np.linspace(t0,t1,6,endpoint=True,dtype="int")
lab = get_xlabels(ticks)
fig[0].set_xticks(ticks)
fig[0].set_xticklabels(lab,fontsize = 12)
fig[0].set_xlabel(
    "Start: {}     Stop {}".format(bkg.fill_start_dt,bkg.fill_end_dt),fontsize = 12)

fig[0].set_xlim(bkg.fill_start,bkg.fill_end)


fig[1].plot(bkg.times,bkg.plusz,color = "royalblue")
fig[1].set_ylabel(r"Beam 1 background [Hz/cm$^2$/10$^{11}$ ]",fontsize = 12)
fig[1].set_xticks(ticks)
fig[1].set_xticklabels(lab)
fig[1].set_xlim(bkg.fill_start,bkg.fill_end)
fig[1].set_xlabel(
    "Start: {}     Stop {}".format(bkg.fill_start_dt,bkg.fill_end_dt),fontsize = 12)
fig[1].grid()

fig[2].plot(bkg.times,bkg.minusz,color = "red")
fig[2].set_ylabel(r"Beam 2 background [Hz/cm$^2$/10$^{11}$ ]",fontsize = 12)
fig[2].set_xticks(ticks)
fig[2].set_xticklabels(lab)
fig[2].set_xlim(bkg.fill_start,bkg.fill_end)
fig[2].set_xlabel(
    "Start: {}     Stop {}".format(bkg.fill_start_dt,bkg.fill_end_dt),fontsize = 12)
fig[2].grid()



plt.subplots_adjust(hspace = 0.25)
plt.savefig("background_graph_1")
```
